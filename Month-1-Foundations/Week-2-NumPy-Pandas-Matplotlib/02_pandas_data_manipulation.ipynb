## 4. Data Selection and Indexing

```python
# Column selection
print("Single column (Series):")
print(df_csv['feature1'].head())
print(f"Type: {type(df_csv['feature1'])}")

print("\nMultiple columns (DataFrame):")
print(df_csv[['feature1', 'feature2']].head())
print(f"Type: {type(df_csv[['feature1', 'feature2']])}")
```

```python
# Row selection with loc (label-based)
print("Select rows 0-4 with loc:")
print(df_csv.loc[0:4, ['id', 'target']])

# Row selection with iloc (position-based)
print("\nSelect rows 0-4 with iloc:")
print(df_csv.iloc[0:5, [0, -1]])

# Select specific rows and columns
print("\nSpecific selection:")
print(df_csv.loc[df_csv['target'] == 'A', ['id', 'feature1', 'target']].head())
```

```python
# Boolean indexing
print("Rows where feature1 > 0:")
print(df_csv[df_csv['feature1'] > 0].head())

print("\nMultiple conditions (AND):")
print(df_csv[(df_csv['feature1'] > 0) & (df_csv['target'] == 'A')].head())

print("\nMultiple conditions (OR):")
print(df_csv[(df_csv['feature1'] > 1) | (df_csv['feature2'] < -1)].head())

# Using query method
print("\nUsing query:")
print(df_csv.query('feature1 > 0 and target == "A"').head())
```

## 5. Data Cleaning

```python
# Create data with missing values
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 8, 9],
    'C': [10, 11, 12, np.nan, 14],
    'D': ['a', 'b', 'c', 'd', 'e']
})

print("DataFrame with missing values:")
print(df_missing)
print(f"\nMissing values:\n{df_missing.isnull().sum()}")
```

```python
# Drop missing values
print("Drop rows with any NaN:")
print(df_missing.dropna())

print("\nDrop columns with any NaN:")
print(df_missing.dropna(axis=1))

print("\nDrop rows where all values are NaN:")
print(df_missing.dropna(how='all'))

print("\nDrop rows with less than 3 non-NaN values:")
print(df_missing.dropna(thresh=3))
```

```python
# Fill missing values
print("Fill with 0:")
print(df_missing.fillna(0))

print("\nFill with mean:")
print(df_missing.fillna(df_missing.mean()))

print("\nForward fill:")
print(df_missing.ffill())

print("\nBackward fill:")
print(df_missing.bfill())

print("\nInterpolate:")
print(df_missing.interpolate())
```

```python
# Remove duplicates
df_dup = pd.DataFrame({
    'A': [1, 1, 2, 2, 3],
    'B': [4, 4, 5, 5, 6],
    'C': [7, 7, 8, 9, 10]
})

print("DataFrame with duplicates:")
print(df_dup)

print("\nCheck for duplicates:")
print(df_dup.duplicated())

print("\nRemove duplicates:")
print(df_dup.drop_duplicates())

print("\nRemove duplicates based on specific columns:")
print(df_dup.drop_duplicates(subset=['A', 'B']))
```

## 6. Data Transformation

```python
# Apply functions
df_transform = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})

print("Original:")
print(df_transform)

# Apply to column
print("\nSquare of column A:")
print(df_transform['A'].apply(lambda x: x**2))

# Apply to DataFrame
print("\nApply to entire DataFrame:")
print(df_transform.apply(np.sqrt))

# Apply along axis
print("\nSum along rows:")
print(df_transform.apply(np.sum, axis=1))
```

```python
# Map values
df_map = pd.DataFrame({
    'grade': ['A', 'B', 'C', 'A', 'B']
})

grade_mapping = {'A': 4.0, 'B': 3.0, 'C': 2.0}
df_map['gpa'] = df_map['grade'].map(grade_mapping)
print("Mapped values:")
print(df_map)

# Replace values
print("\nReplace A with A+:")
print(df_map.replace('A', 'A+'))
```

```python
# Create new columns
df_new = pd.DataFrame({
    'price': [100, 200, 300],
    'quantity': [2, 3, 4]
})

# Simple calculation
df_new['total'] = df_new['price'] * df_new['quantity']

# Using assign
df_new = df_new.assign(
    discount=df_new['total'] * 0.1,
    final_price=lambda x: x['total'] - x['discount']
)

print("DataFrame with new columns:")
print(df_new)
```

```python
# Binning
ages = pd.Series([5, 15, 25, 35, 45, 55, 65, 75])

# Cut into bins
age_groups = pd.cut(ages, bins=[0, 18, 35, 60, 100], 
                    labels=['Child', 'Young Adult', 'Adult', 'Senior'])
print("Age groups:")
print(age_groups)

# Quantile-based binning
quartiles = pd.qcut(ages, q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
print("\nQuartiles:")
print(quartiles)
```

## 7. GroupBy Operations

```python
# Create sample data
df_group = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'subcategory': ['X', 'X', 'Y', 'Y', 'X', 'X', 'Y', 'Y'],
    'value1': [10, 15, 20, 25, 30, 35, 40, 45],
    'value2': [100, 150, 200, 250, 300, 350, 400, 450]
})

print("Original DataFrame:")
print(df_group)
```

```python
# Basic groupby
print("Group by category and compute mean:")
print(df_group.groupby('category').mean())

print("\nGroup by category and compute sum:")
print(df_group.groupby('category').sum())

print("\nMultiple aggregations:")
print(df_group.groupby('category').agg(['mean', 'sum', 'std']))
```

```python
# Multiple groupby columns
print("Group by multiple columns:")
print(df_group.groupby(['category', 'subcategory']).mean())

# Different aggregations for different columns
print("\nDifferent aggregations:")
print(df_group.groupby('category').agg({
    'value1': ['mean', 'sum'],
    'value2': ['min', 'max']
}))
```

```python
# Custom aggregation functions
def range_func(x):
    return x.max() - x.min()

print("Custom aggregation:")
print(df_group.groupby('category').agg({
    'value1': [np.mean, range_func],
    'value2': [np.median, 'std']
}))

# Transform (returns same shape as input)
print("\nTransform (normalize within groups):")
df_group['value1_normalized'] = df_group.groupby('category')['value1'].transform(
    lambda x: (x - x.mean()) / x.std()
)
print(df_group)
```

## 8. Merging and Joining

```python
# Create sample dataframes
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value1': [1, 2, 3, 4]
})

df2 = pd.DataFrame({
    'key': ['B', 'D', 'E', 'F'],
    'value2': [5, 6, 7, 8]
})

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
```

```python
# Inner join (intersection)
print("Inner join:")
print(pd.merge(df1, df2, on='key', how='inner'))

# Left join
print("\nLeft join:")
print(pd.merge(df1, df2, on='key', how='left'))

# Right join
print("\nRight join:")
print(pd.merge(df1, df2, on='key', how='right'))

# Outer join (union)
print("\nOuter join:")
print(pd.merge(df1, df2, on='key', how='outer'))
```

```python
# Concatenate DataFrames
df_a = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_b = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print("Concatenate vertically (rows):")
print(pd.concat([df_a, df_b], axis=0, ignore_index=True))

print("\nConcatenate horizontally (columns):")
print(pd.concat([df_a, df_b], axis=1))
```

## 9. Pivot Tables and Reshaping

```python
# Create sample data
df_pivot = pd.DataFrame({
    'date': ['2024-01', '2024-01', '2024-02', '2024-02'] * 2,
    'category': ['A', 'B', 'A', 'B'] * 2,
    'region': ['North', 'North', 'North', 'North', 'South', 'South', 'South', 'South'],
    'sales': [100, 150, 120, 180, 90, 140, 110, 170]
})

print("Original data:")
print(df_pivot)
```

```python
# Pivot table
print("Pivot table:")
pivot = df_pivot.pivot_table(
    values='sales',
    index='date',
    columns='category',
    aggfunc='sum'
)
print(pivot)

print("\nPivot with multiple aggregations:")
pivot_multi = df_pivot.pivot_table(
    values='sales',
    index='date',
    columns='category',
    aggfunc=['sum', 'mean']
)
print(pivot_multi)
```

```python
# Melt (unpivot)
df_wide = pd.DataFrame({
    'id': [1, 2, 3],
    'Jan': [100, 110, 120],
    'Feb': [105, 115, 125],
    'Mar': [110, 120, 130]
})

print("Wide format:")
print(df_wide)

print("\nLong format (melted):")
df_long = df_wide.melt(id_vars=['id'], 
                       var_name='month', 
                       value_name='sales')
print(df_long)
```

## 10. Time Series Operations

```python
# Create time series data
dates = pd.date_range('2024-01-01', periods=100, freq='D')
ts_data = pd.DataFrame({
    'date': dates,
    'value': np.random.randn(100).cumsum()
})

ts_data.set_index('date', inplace=True)
print("Time series data:")
print(ts_data.head(10))
```

```python
# Date operations
ts_data['year'] = ts_data.index.year
ts_data['month'] = ts_data.index.month
ts_data['day'] = ts_data.index.day
ts_data['dayofweek'] = ts_data.index.dayofweek

print("With date components:")
print(ts_data.head())
```

```python
# Resampling
print("Weekly mean:")
print(ts_data['value'].resample('W').mean().head())

print("\nMonthly sum:")
print(ts_data['value'].resample('M').sum())

# Rolling window
print("\n7-day rolling mean:")
ts_data['rolling_mean'] = ts_data['value'].rolling(window=7).mean()
print(ts_data[['value', 'rolling_mean']].head(10))
```

## 11. String Operations

```python
# String methods
df_str = pd.DataFrame({
    'name': ['Alice Smith', 'Bob Jones', 'Charlie Brown'],
    'email': ['alice@example.com', 'bob@example.com', 'charlie@example.com']
})

print("Original:")
print(df_str)

# String operations
df_str['first_name'] = df_str['name'].str.split().str[0]
df_str['last_name'] = df_str['name'].str.split().str[1]
df_str['name_upper'] = df_str['name'].str.upper()
df_str['name_length'] = df_str['name'].str.len()
df_str['domain'] = df_str['email'].str.split('@').str[1]

print("\nWith string operations:")
print(df_str)
```

## 12. Performance Tips

```python
import time

# Create large DataFrame
large_df = pd.DataFrame({
    'A': np.random.randn(100000),
    'B': np.random.randn(100000),
    'C': np.random.choice(['X', 'Y', 'Z'], 100000)
})

# Vectorized operations (fast)
start = time.time()
result_vec = large_df['A'] * 2 + large_df['B']
vec_time = time.time() - start

# Iterrows (slow - avoid!)
start = time.time()
result_iter = []
for idx, row in large_df.head(1000).iterrows():  # Only 1000 rows for demo
    result_iter.append(row['A'] * 2 + row['B'])
iter_time = time.time() - start

print(f"Vectorized time: {vec_time:.4f}s")
print(f"Iterrows time (1000 rows): {iter_time:.4f}s")
print(f"Vectorized is {iter_time / vec_time:.0f}x faster!")

print("\n⚠️ Key Performance Tips:")
print("1. Use vectorized operations instead of loops")
print("2. Use categorical dtype for string columns with few unique values")
print("3. Read only needed columns with usecols")
print("4. Use chunking for very large files")
print("5. Avoid iterrows() - use apply() or vectorization")
```
