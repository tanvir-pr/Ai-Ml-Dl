# Data Visualization with Matplotlib and Seaborn

## Overview
Visualization is crucial for understanding data and communicating insights. This notebook covers:
- Matplotlib fundamentals
- Common plot types
- Customization and styling
- Seaborn for statistical plots
- Best practices for ML visualizations

**Time:** 4-5 hours  
**Prerequisites:** NumPy, Pandas basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style (use default if seaborn style not available)
try:
    plt.style.use('seaborn-v0_8-darkgrid')
except OSError:
    try:
        plt.style.use('seaborn-darkgrid')
    except OSError:
        pass  # use default style
sns.set_palette("husl")

# For inline plots in Jupyter
%matplotlib inline

print(f"Matplotlib version: {plt.matplotlib.__version__}")
print(f"Seaborn version: {sns.__version__}")
```

## 1. Matplotlib Basics

```python
# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Sine Wave')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Multiple lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='cos(x)', linewidth=2, linestyle='--')
plt.plot(x, y3, label='sin(x)*cos(x)', linewidth=2, linestyle=':')
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('Trigonometric Functions', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Scatter plot
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, s=50, c=y, cmap='viridis')
plt.colorbar(label='Y value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Color Mapping')
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Common Plot Types

```python
# Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color='skyblue', edgecolor='navy', alpha=0.7)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True, alpha=0.3)

# Histogram with KDE
plt.subplot(1, 2, 2)
plt.hist(data, bins=30, density=True, edgecolor='black', alpha=0.7, label='Histogram')
from scipy import stats
x_range = np.linspace(data.min(), data.max(), 100)
kde = stats.gaussian_kde(data)
plt.plot(x_range, kde(x_range), 'r-', linewidth=2, label='KDE')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with KDE')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(10, 6))
bp = plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'],
                 patch_artist=True, notch=True)

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Values')
plt.title('Box Plot Comparison')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

```python
# Pie chart
sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

plt.figure(figsize=(10, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pie Chart Example', fontsize=14, fontweight='bold')
plt.show()
```

## 3. Subplots and Layouts

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Line plot
x = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Sine Wave')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Scatter plot
x = np.random.randn(50)
y = np.random.randn(50)
axes[0, 1].scatter(x, y, alpha=0.6)
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Histogram
data = np.random.randn(1000)
axes[1, 0].hist(data, bins=30, edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Histogram')
axes[1, 0].grid(True, alpha=0.3)

# Plot 4: Bar plot
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
axes[1, 1].bar(categories, values, color='skyblue', edgecolor='navy')
axes[1, 1].set_title('Bar Chart')
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# GridSpec for complex layouts
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(12, 8))
gs = GridSpec(3, 3, figure=fig)

# Large plot spanning 2x2
ax1 = fig.add_subplot(gs[0:2, 0:2])
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), 'b-', linewidth=2)
ax1.set_title('Main Plot')
ax1.grid(True, alpha=0.3)

# Smaller plots
ax2 = fig.add_subplot(gs[0, 2])
ax2.hist(np.random.randn(100), bins=20)
ax2.set_title('Hist 1')

ax3 = fig.add_subplot(gs[1, 2])
ax3.scatter(np.random.randn(50), np.random.randn(50))
ax3.set_title('Scatter')

ax4 = fig.add_subplot(gs[2, :])
ax4.plot(np.random.randn(100).cumsum())
ax4.set_title('Bottom Plot')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Customization and Styling

```python
# Available styles
print("Available styles:")
print(plt.style.available[:10])  # Show first 10

# Compare different styles
styles = ['default', 'ggplot', 'bmh', 'classic']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

x = np.linspace(0, 10, 100)
y = np.sin(x)

for ax, style in zip(axes, styles):
    with plt.style.context(style):
        ax.plot(x, y, linewidth=2)
        ax.set_title(f'Style: {style}')
        ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Custom colors and markers
x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12, 6))

# Line with markers
plt.plot(x, y1, color='#FF6B6B', marker='o', markersize=8, 
         linestyle='-', linewidth=2, label='sin(x)',
         markerfacecolor='white', markeredgewidth=2, markeredgecolor='#FF6B6B')

plt.plot(x, y2, color='#4ECDC4', marker='s', markersize=8,
         linestyle='--', linewidth=2, label='cos(x)',
         markerfacecolor='white', markeredgewidth=2, markeredgecolor='#4ECDC4')

plt.xlabel('X', fontsize=14, fontweight='bold')
plt.ylabel('Y', fontsize=14, fontweight='bold')
plt.title('Custom Styling Example', fontsize=16, fontweight='bold')
plt.legend(fontsize=12, loc='upper right', framealpha=0.9)
plt.grid(True, alpha=0.3, linestyle=':', linewidth=1)
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.show()
```

```python
# Annotations and text
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y, linewidth=2)

# Add annotations
plt.annotate('Maximum', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.3),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=12, color='red')

plt.annotate('Minimum', xy=(3*np.pi/2, -1), xytext=(3*np.pi/2 + 1, -1.3),
             arrowprops=dict(arrowstyle='->', color='blue', lw=2),
             fontsize=12, color='blue')

# Add text
plt.text(5, 0.5, 'Sine Wave', fontsize=14, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot with Annotations')
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. Seaborn Statistical Plots

```python
# Create sample dataset
np.random.seed(42)
df = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'category': np.random.choice(['A', 'B', 'C'], 100),
    'value': np.random.randint(1, 100, 100)
})

print("Sample data:")
print(df.head())
```

```python
# Distribution plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram with KDE
sns.histplot(data=df, x='x', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Histogram with KDE')

# KDE plot
sns.kdeplot(data=df, x='x', hue='category', ax=axes[0, 1])
axes[0, 1].set_title('KDE Plot by Category')

# Box plot
sns.boxplot(data=df, x='category', y='value', ax=axes[1, 0])
axes[1, 0].set_title('Box Plot')

# Violin plot
sns.violinplot(data=df, x='category', y='value', ax=axes[1, 1])
axes[1, 1].set_title('Violin Plot')

plt.tight_layout()
plt.show()
```

```python
# Scatter plots with regression
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Scatter plot
sns.scatterplot(data=df, x='x', y='y', hue='category', 
                size='value', sizes=(20, 200), ax=axes[0])
axes[0].set_title('Scatter Plot with Size and Color')

# Regression plot
sns.regplot(data=df, x='x', y='y', ax=axes[1])
axes[1].set_title('Scatter Plot with Regression Line')

plt.tight_layout()
plt.show()
```

```python
# Pair plot
# Create dataset with multiple features
iris_like = pd.DataFrame({
    'feature1': np.random.randn(100),
    'feature2': np.random.randn(100),
    'feature3': np.random.randn(100),
    'class': np.random.choice(['A', 'B', 'C'], 100)
})

# Pair plot
sns.pairplot(iris_like, hue='class', diag_kind='kde', corner=True)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

```python
# Heatmap
# Create correlation matrix
corr_data = pd.DataFrame(np.random.randn(100, 5), 
                         columns=['A', 'B', 'C', 'D', 'E'])
correlation = corr_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Joint plot
sns.jointplot(data=df, x='x', y='y', kind='hex', height=8)
plt.suptitle('Joint Plot with Hexbin', y=1.02)
plt.show()

# With regression
sns.jointplot(data=df, x='x', y='y', kind='reg', height=8)
plt.suptitle('Joint Plot with Regression', y=1.02)
plt.show()
```

## 6. ML-Specific Visualizations

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

# Simulated predictions
y_true = np.random.choice([0, 1, 2], 100)
y_pred = np.random.choice([0, 1, 2], 100)

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Class 0', 'Class 1', 'Class 2'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Learning Curves
epochs = np.arange(1, 51)
train_loss = np.exp(-epochs/10) + np.random.randn(50) * 0.05
val_loss = np.exp(-epochs/10) + 0.1 + np.random.randn(50) * 0.05
train_acc = 1 - np.exp(-epochs/10) + np.random.randn(50) * 0.02
val_acc = 1 - np.exp(-epochs/10) - 0.05 + np.random.randn(50) * 0.02

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Loss curves
ax1.plot(epochs, train_loss, label='Training Loss', linewidth=2)
ax1.plot(epochs, val_loss, label='Validation Loss', linewidth=2)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.set_title('Training and Validation Loss')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Accuracy curves
ax2.plot(epochs, train_acc, label='Training Accuracy', linewidth=2)
ax2.plot(epochs, val_acc, label='Validation Accuracy', linewidth=2)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.set_title('Training and Validation Accuracy')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ROC Curve
from sklearn.metrics import roc_curve, auc

# Simulated probabilities and labels
y_true = np.random.choice([0, 1], 100)
y_scores = np.random.rand(100)

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, 
         label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Feature Importance
features = ['Feature A', 'Feature B', 'Feature C', 'Feature D', 'Feature E']
importance = np.array([0.25, 0.20, 0.18, 0.22, 0.15])

# Sort by importance
indices = np.argsort(importance)

plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), importance[indices], color='skyblue', edgecolor='navy')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.grid(True, alpha=0.3, axis='x')

# Add values on bars
for i, v in enumerate(importance[indices]):
    plt.text(v + 0.01, i, f'{v:.2f}', va='center')

plt.tight_layout()
plt.show()
```

## 8. Best Practices

```python
# Good vs Bad visualization
data = np.random.randn(100)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bad: No labels, no title, poor colors
ax1.hist(data, bins=20, color='red')
ax1.set_title('❌ Bad Visualization')

# Good: Clear labels, title, appropriate colors, grid
ax2.hist(data, bins=20, edgecolor='black', alpha=0.7, color='skyblue')
ax2.set_xlabel('Value', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.set_title('✅ Good Visualization', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\n📊 Visualization Best Practices:")
print("1. Always label axes")
print("2. Include informative titles")
print("3. Use appropriate colors (colorblind-friendly)")
print("4. Add grid for easier reading")
print("5. Use legends when multiple series")
print("6. Choose appropriate plot type for data")
print("7. Keep it simple - avoid chart junk")
print("8. Use consistent styling")
print("9. Consider your audience")
print("10. Save in appropriate format (vector for publications)")
```
