## 1. Python Data Structures Review

```python
# Lists - ordered, mutable collections
data_points = [1.5, 2.3, 3.7, 4.2, 5.1]
print(f"Original list: {data_points}")
print(f"First element: {data_points[0]}")
print(f"Last element: {data_points[-1]}")
print(f"Slice [1:3]: {data_points[1:3]}")

# List operations
data_points.append(6.8)
data_points.extend([7.2, 8.5])
print(f"After append and extend: {data_points}")
```

```python
# Dictionaries - key-value pairs (like JSON)
model_config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 100,
    'optimizer': 'adam'
}

print(f"Learning rate: {model_config['learning_rate']}")
print(f"All keys: {model_config.keys()}")
print(f"All values: {model_config.values()}")

# Safe access with get()
momentum = model_config.get('momentum', 0.9)  # default value if key doesn't exist
print(f"Momentum: {momentum}")
```

```python
# Tuples - immutable sequences
image_shape = (224, 224, 3)  # height, width, channels
print(f"Image shape: {image_shape}")
height, width, channels = image_shape  # unpacking
print(f"Height: {height}, Width: {width}, Channels: {channels}")

# Sets - unordered collections of unique elements
features_a = {'age', 'income', 'education', 'location'}
features_b = {'age', 'income', 'gender', 'occupation'}

print(f"Common features: {features_a & features_b}")
print(f"All features: {features_a | features_b}")
print(f"Features only in A: {features_a - features_b}")
```

## 2. List Comprehensions and Generators

```python
# Traditional loop
squares = []
for i in range(10):
    squares.append(i ** 2)
print(f"Squares (loop): {squares}")

# List comprehension - more Pythonic
squares = [i ** 2 for i in range(10)]
print(f"Squares (comprehension): {squares}")

# With condition
even_squares = [i ** 2 for i in range(10) if i % 2 == 0]
print(f"Even squares: {even_squares}")

# Nested comprehension for matrix
matrix = [[i * j for j in range(3)] for i in range(3)]
print(f"Matrix:\n{matrix}")
```

```python
# Dictionary comprehension
feature_means = {'feature_' + str(i): i * 10.5 for i in range(5)}
print(f"Feature means: {feature_means}")

# Generator expression (memory efficient for large datasets)
large_dataset = (i ** 2 for i in range(1000000))  # doesn't create list in memory
print(f"Generator object: {large_dataset}")
print(f"First 5 values: {[next(large_dataset) for _ in range(5)]}")
```

## 3. Functions and Lambda Expressions

```python
# Regular function
def normalize(x, mean, std):
    """Normalize a value using z-score normalization."""
    return (x - mean) / std

# Function with default arguments
def train_model(data, learning_rate=0.001, epochs=100, verbose=True):
    """Simulate model training."""
    if verbose:
        print(f"Training with lr={learning_rate} for {epochs} epochs")
    return {"loss": 0.05, "accuracy": 0.95}

result = train_model(None, learning_rate=0.01, epochs=50)
print(f"Training result: {result}")
```

```python
# Lambda functions - anonymous functions
square = lambda x: x ** 2
print(f"Square of 5: {square(5)}")

# Using lambda with map, filter, reduce
numbers = [1, 2, 3, 4, 5]

# Map - apply function to each element
squared = list(map(lambda x: x ** 2, numbers))
print(f"Squared: {squared}")

# Filter - keep elements that satisfy condition
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {evens}")

# Reduce - accumulate values
from functools import reduce
product = reduce(lambda x, y: x * y, numbers)
print(f"Product: {product}")
```

```python
# Higher-order functions
def apply_activation(activation_fn):
    """Returns a function that applies activation to input."""
    def activate(x):
        return activation_fn(x)
    return activate

# ReLU activation
relu = apply_activation(lambda x: max(0, x))
print(f"ReLU(-5): {relu(-5)}")
print(f"ReLU(5): {relu(5)}")

# Sigmoid activation
import math
sigmoid = apply_activation(lambda x: 1 / (1 + math.exp(-x)))
print(f"Sigmoid(0): {sigmoid(0)}")
print(f"Sigmoid(2): {sigmoid(2):.4f}")
```

## 4. Object-Oriented Programming (OOP)

```python
# Basic class definition
class DataPoint:
    """Represents a single data point with features and label."""
    
    def __init__(self, features, label):
        self.features = features
        self.label = label
    
    def __repr__(self):
        return f"DataPoint(features={self.features}, label={self.label})"
    
    def get_feature_count(self):
        return len(self.features)

# Create instances
point1 = DataPoint([1.5, 2.3, 3.7], 'positive')
point2 = DataPoint([0.8, 1.2, 2.1], 'negative')

print(point1)
print(f"Feature count: {point1.get_feature_count()}")
```

```python
# More complex class with inheritance
class Model:
    """Base class for ML models."""
    
    def __init__(self, name):
        self.name = name
        self.is_trained = False
    
    def train(self, X, y):
        """Train the model."""
        raise NotImplementedError("Subclass must implement train method")
    
    def predict(self, X):
        """Make predictions."""
        if not self.is_trained:
            raise ValueError("Model must be trained before prediction")
        raise NotImplementedError("Subclass must implement predict method")

class LinearRegression(Model):
    """Simple linear regression model."""
    
    def __init__(self):
        super().__init__("LinearRegression")
        self.weights = None
        self.bias = None
    
    def train(self, X, y):
        """Train using simple approach (placeholder)."""
        print(f"Training {self.name}...")
        self.weights = [0.5] * len(X[0]) if X else []
        self.bias = 0.1
        self.is_trained = True
        print("Training complete!")
    
    def predict(self, X):
        """Make predictions."""
        super().predict(X)  # check if trained
        predictions = [sum(w * x for w, x in zip(self.weights, features)) + self.bias 
                      for features in X]
        return predictions

# Use the model
model = LinearRegression()
X_train = [[1, 2], [2, 3], [3, 4]]
y_train = [3, 5, 7]

model.train(X_train, y_train)
predictions = model.predict([[4, 5], [5, 6]])
print(f"Predictions: {predictions}")
```

```python
# Class with properties and special methods
class Dataset:
    """Container for ML dataset."""
    
    def __init__(self, X, y):
        self._X = X
        self._y = y
    
    @property
    def shape(self):
        """Return dataset shape."""
        return (len(self._X), len(self._X[0]) if self._X else 0)
    
    @property
    def X(self):
        return self._X
    
    @property
    def y(self):
        return self._y
    
    def __len__(self):
        return len(self._X)
    
    def __getitem__(self, idx):
        return self._X[idx], self._y[idx]
    
    def __repr__(self):
        return f"Dataset(samples={len(self)}, features={self.shape[1]})"

# Create and use dataset
dataset = Dataset([[1, 2], [3, 4], [5, 6]], [0, 1, 0])
print(dataset)
print(f"Shape: {dataset.shape}")
print(f"First sample: {dataset[0]}")
print(f"Length: {len(dataset)}")
```

## 5. Decorators and Context Managers

```python
# Simple decorator
import time

def timer(func):
    """Decorator to measure function execution time."""
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def train_large_model():
    """Simulate training."""
    time.sleep(0.1)
    return "Model trained"

result = train_large_model()
```

```python
# Context manager for resource management
class DataLoader:
    """Context manager for loading data."""
    
    def __init__(self, filename):
        self.filename = filename
        self.data = None
    
    def __enter__(self):
        print(f"Loading data from {self.filename}...")
        self.data = [1, 2, 3, 4, 5]  # simulate loading
        return self.data
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("Cleaning up resources...")
        self.data = None
        return False

# Use context manager
with DataLoader("train.csv") as data:
    print(f"Processing {len(data)} samples")
    print(f"Data: {data}")

print("Context exited")
```

## 6. Exception Handling

```python
# Basic exception handling
def safe_divide(a, b):
    """Safely divide two numbers."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
        return None
    except TypeError:
        print("Error: Invalid types for division")
        return None
    finally:
        print("Division operation completed")

print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

```python
# Custom exceptions
class ModelNotTrainedError(Exception):
    """Raised when trying to use untrained model."""
    pass

class InsufficientDataError(Exception):
    """Raised when dataset is too small."""
    pass

def validate_dataset(X, y, min_samples=10):
    """Validate dataset before training."""
    if len(X) != len(y):
        raise ValueError("X and y must have same length")
    if len(X) < min_samples:
        raise InsufficientDataError(f"Need at least {min_samples} samples, got {len(X)}")
    print("Dataset validation passed")

# Test validation
try:
    validate_dataset([1, 2, 3], [0, 1, 0], min_samples=5)
except InsufficientDataError as e:
    print(f"Validation failed: {e}")
```
