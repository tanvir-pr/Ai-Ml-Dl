# Hypothesis Testing for Machine Learning

## Overview
- Null and alternative hypotheses
- p-values and significance levels
- t-tests (one-sample, two-sample)
- Chi-square test
- Confidence intervals
- Type I and Type II errors

**Time:** 3-4 hours

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(42)
```

## 1. Hypothesis Testing Basics

```python
# H0: Null hypothesis (no effect)
# H1: Alternative hypothesis (there is an effect)
# p-value: Probability of observing data if H0 is true
# alpha: Significance level (usually 0.05)
# If p < alpha: Reject H0 (statistically significant)
# If p >= alpha: Fail to reject H0
```

## 2. One-Sample t-test

```python
# Test if sample mean differs from population mean
# H0: mean = 0
sample = np.random.randn(30) + 0.5  # Mean ~0.5
statistic, p_value = stats.ttest_1samp(sample, 0)
print(f't-statistic: {statistic:.4f}')
print(f'p-value: {p_value:.4f}')
print(f'Reject H0 (mean=0)? {p_value < 0.05}')
```

## 3. Two-Sample t-test

```python
# Test if two groups have different means
# H0: mean1 = mean2
group1 = np.random.randn(30) + 0
group2 = np.random.randn(30) + 1  # Different mean

statistic, p_value = stats.ttest_ind(group1, group2)
print(f't-statistic: {statistic:.4f}')
print(f'p-value: {p_value:.6f}')
print(f'Groups different? {p_value < 0.05}')
```

## 4. Chi-Square Test

```python
# Test independence of categorical variables
# Example: Is outcome independent of group?
observed = np.array([[10, 20], [20, 50]])  # 2x2 contingency table
chi2, p_value, dof, expected = stats.chi2_contingency(observed)
print(f'Chi-square: {chi2:.4f}')
print(f'p-value: {p_value:.4f}')
print(f'Variables independent? {p_value >= 0.05}')
```

## 5. Confidence Intervals

```python
# 95% CI for mean
sample = np.random.randn(100)
mean = np.mean(sample)
se = stats.sem(sample)
ci_95 = stats.t.interval(0.95, len(sample)-1, loc=mean, scale=se)
print(f'Sample mean: {mean:.4f}')
print(f'95% CI: [{ci_95[0]:.4f}, {ci_95[1]:.4f}]')
```

## 6. Summary

- **t-test**: Compare means (parametric)
- **Chi-square**: Categorical independence
- **p-value < 0.05**: Statistically significant
- **Confidence intervals**: Range of plausible values
- **Type I error**: False positive (reject true H0)
- **Type II error**: False negative (fail to reject false H0)
