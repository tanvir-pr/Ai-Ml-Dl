```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## 1. Basic Probability

```python
# P(A) = favorable outcomes / total outcomes
# Example: Probability of rolling 6 on fair die
p_six = 1/6
print(f'P(rolling 6) = {p_six}')

# Complement: P(not A) = 1 - P(A)
p_not_six = 1 - p_six
print(f'P(not 6) = {p_not_six}')

# Union: P(A or B) = P(A) + P(B) - P(A and B)
# Intersection: P(A and B) = P(A) * P(B) if independent
```

## 2. Conditional Probability

```python
# P(A|B) = P(A and B) / P(B)
# Example: Disease testing
# P(disease) = 0.01, P(positive|disease) = 0.99, P(positive|no disease) = 0.05

p_disease = 0.01
p_positive_given_disease = 0.99
p_positive_given_no_disease = 0.05

# P(positive) = P(positive and disease) + P(positive and no disease)
p_positive = p_positive_given_disease * p_disease + p_positive_given_no_disease * (1 - p_disease)
print(f'P(positive test) = {p_positive:.4f}')

# P(disease|positive) - we need Bayes for this
```

## 3. Bayes' Theorem

```python
# Bayes: P(A|B) = P(B|A) * P(A) / P(B)
# P(disease|positive) = P(positive|disease) * P(disease) / P(positive)

p_disease_given_positive = (p_positive_given_disease * p_disease) / p_positive
print(f'P(disease | positive test) = {p_disease_given_positive:.4f}')
print('\nEven with positive test, only ~17% chance of having disease!')
print('This is due to low base rate (1% prevalence).')
```

## 4. Random Variables and Expectation

```python
# E[X] = sum of x * P(X=x)
# For die: E[X] = (1+2+3+4+5+6)/6 = 3.5
die_outcomes = np.arange(1, 7)
e_die = np.mean(die_outcomes)
print(f'E[die roll] = {e_die}')

# Variance: Var(X) = E[X^2] - E[X]^2
var_die = np.var(die_outcomes)
print(f'Var[die roll] = {var_die}')
```

## 5. Law of Large Numbers

```python
# Sample mean converges to population mean as n increases
means = []
for n in [10, 100, 1000, 10000, 100000]:
    samples = np.random.randn(n)
    means.append(np.mean(samples))

plt.figure(figsize=(10, 5))
plt.plot([10, 100, 1000, 10000, 100000], means, 'bo-')
plt.axhline(y=0, color='r', linestyle='--')
plt.xscale('log')
plt.xlabel('Sample size')
plt.ylabel('Sample mean')
plt.title('Law of Large Numbers')
plt.show()
```

## 6. Summary

- **Conditional probability** is fundamental for ML (e.g., P(y|x))
- **Bayes' theorem** underlies Naive Bayes, spam filters
- **Expectation** used in loss functions, optimization
- **Law of large numbers** justifies using sample statistics
