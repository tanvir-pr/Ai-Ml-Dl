# Simple Linear Regression

## Overview
- Theory: y = β₀ + β₁x
- Cost function (MSE)
- Normal equation (closed-form solution)
- Implementation from scratch
- Using scikit-learn

**Time:** 2-3 hours

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(42)
```

```python
# Generate data: y = 4 + 3x + noise
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y, alpha=0.5)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression Data')
plt.show()
```

```python
# Normal equation: β = (X^T X)^(-1) X^T y
X_b = np.c_[np.ones((100, 1)), X]
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print(f'Intercept (β0): {beta[0,0]:.4f}')
print(f'Slope (β1): {beta[1,0]:.4f}')
print('True values: β0=4, β1=3')
```

```python
# Predict and plot
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b @ beta

plt.scatter(X, y, alpha=0.5)
plt.plot(X_new, y_pred, 'r-', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Fitted Regression Line')
plt.show()
```

```python
# Using scikit-learn
model = LinearRegression()
model.fit(X, y)
print(f'Intercept: {model.intercept_[0]:.4f}')
print(f'Slope: {model.coef_[0,0]:.4f}')
y_pred_sk = model.predict(X)
print(f'MSE: {mean_squared_error(y, y_pred_sk):.4f}')
print(f'R²: {r2_score(y, y_pred_sk):.4f}')
```
