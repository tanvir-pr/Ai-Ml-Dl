# Gradient Boosting: XGBoost, LightGBM, CatBoost

## Overview
- Boosting: sequential correction of errors
- XGBoost: most popular
- LightGBM: faster, less memory
- CatBoost: handles categorical natively

**Time:** 2-3 hours

```python
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score

X, y = load_iris(return_X_y=True)
model = xgb.XGBClassifier(n_estimators=100, max_depth=3, random_state=42)
scores = cross_val_score(model, X, y, cv=5)
print('XGBoost CV accuracy:', scores.mean(), '+/-', scores.std())
```

## 2. If XGBoost is not installed

Use **sklearn** `GradientBoostingClassifier` as a fallback (`pip install xgboost` for the cell above).

```python
try:
    import xgboost  # noqa: F401
    print('XGBoost is available; primary example is in the previous cell.')
except ImportError:
    from sklearn.ensemble import GradientBoostingClassifier
    gbc = GradientBoostingClassifier(n_estimators=100, max_depth=3, random_state=42)
    from sklearn.datasets import load_iris
    from sklearn.model_selection import cross_val_score
    Xi, yi = load_iris(return_X_y=True)
    sc = cross_val_score(gbc, Xi, yi, cv=5)
    print('sklearn GradientBoostingClassifier CV mean:', sc.mean())
```
