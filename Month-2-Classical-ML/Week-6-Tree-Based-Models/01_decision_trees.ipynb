# Decision Trees

## Overview
- CART algorithm
- Entropy and Gini impurity
- Splitting criteria
- Overfitting and pruning

**Time:** 2-3 hours

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

X, y = load_iris(return_X_y=True)
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X, y)
plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=load_iris().feature_names, class_names=load_iris().target_names, filled=True)
plt.show()
print('Accuracy:', model.score(X, y))
```

## 2. Impurity measures

**Entropy** at a node: $H = -\sum_k p_k \log_2 p_k$ with $p_k$ = fraction of class $k$.  
**Gini:** $G = 1 - \sum_k p_k^2$.  
Lower impurity → purer node. Splits maximize **information gain** (parent impurity − weighted child impurity).

```python
from sklearn.model_selection import cross_val_score

depths = range(1, 15)
cv_means = []
for d in depths:
    clf = DecisionTreeClassifier(max_depth=d, random_state=42)
    cv_means.append(cross_val_score(clf, X, y, cv=5).mean())

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(list(depths), cv_means, 'o-')
plt.xlabel('max_depth')
plt.ylabel('5-fold CV accuracy')
plt.title('Depth vs generalization (Iris)')
plt.grid(True, alpha=0.3)
plt.show()
```
