# Random Forests

## Overview
- Bagging (Bootstrap Aggregating)
- Feature subset at each split
- Reduces variance
- Feature importance

**Time:** 2 hours

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
print('Accuracy:', rf.score(X, y))
print('Feature importances:', dict(zip(load_iris().feature_names, rf.feature_importances_)))
```

## 2. Out-of-bag (OOB) error

Each tree trains on a bootstrap sample; **OOB** estimates generalization using samples **not** in that tree's bag. Set `oob_score=True`.

```python
rf_oob = RandomForestClassifier(
    n_estimators=100, oob_score=True, random_state=42
)
rf_oob.fit(X, y)
print('OOB score (accuracy estimate):', rf_oob.oob_score_)
```
