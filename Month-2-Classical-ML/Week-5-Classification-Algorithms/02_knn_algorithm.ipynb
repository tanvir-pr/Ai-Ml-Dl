# K-Nearest Neighbors (KNN)

## Overview
- Instance-based learning
- Distance metrics (Euclidean, Manhattan)
- Choosing k
- Pros/cons

**Time:** 2 hours

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score

X, y = load_iris(return_X_y=True)

# Find best k
scores = []
for k in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, y, cv=5).mean()
    scores.append(score)

print('Best k:', np.argmax(scores) + 1)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
```

## 2. Distance and memory

- **Euclidean** (default `p=2`): straight-line distance in feature space.
- **Manhattan** (`p=1`): sum of absolute differences — sometimes better for high dimensions.
- **Feature scaling** is required: KNN uses raw distances; one large-scale feature dominates otherwise.

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay

X_s = StandardScaler().fit_transform(X)
best_k = int(np.argmax(scores)) + 1
knn_viz = KNeighborsClassifier(n_neighbors=best_k)
knn_viz.fit(X_s[:, :2], y)  # first 2 features for 2D plot if iris

fig, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    knn_viz, X_s[:, :2], response_method='predict', alpha=0.3, ax=ax
)
ax.scatter(X_s[:, 0], X_s[:, 1], c=y, edgecolors='k', s=30)
ax.set_title(f'KNN boundaries (k={best_k}, first 2 features)')
plt.show()
```
