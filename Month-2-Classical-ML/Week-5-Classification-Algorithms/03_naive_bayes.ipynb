```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

# Gaussian NB for continuous features
gnb = GaussianNB()
gnb.fit(X, y)
print('Gaussian NB accuracy:', gnb.score(X, y))

# Multinomial NB for count data (e.g., text)
# Use for document classification with word counts
```

## 2. Bayes' rule for classification

**Naive Bayes** assumes features are **independent** given the class:

$$P(y|x) \propto P(y) \prod_i P(x_i|y)$$

**GaussianNB:** each $P(x_i|y)$ is a normal distribution (good for continuous tabular data).  
**MultinomialNB:** counts or discrete features (common with text word counts).  
**BernoulliNB:** binary features.

```python
# Class prior probabilities (empirical)
print('Class priors:', gnb.class_prior_)
print('Means (per class, per feature) shape:', gnb.theta_.shape)

# Multinomial example with small synthetic count data
import numpy as np
X_counts = np.random.randint(0, 10, size=(100, 4))
y_bin = np.random.randint(0, 2, size=100)
mnb = MultinomialNB()
mnb.fit(X_counts, y_bin)
print('MultinomialNB accuracy (train):', mnb.score(X_counts, y_bin))
```
