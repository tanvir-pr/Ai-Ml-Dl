# Support Vector Machines (SVM)

## Overview
- Maximum margin classifier
- Kernel trick (linear, RBF, polynomial)
- C parameter and slack
- SVC for classification

**Time:** 2-3 hours

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler

X, y = make_classification(n_samples=200, n_features=2, random_state=42)
X = StandardScaler().fit_transform(X)

# Linear kernel
svm_linear = SVC(kernel='linear').fit(X, y)
print('Linear SVM score:', svm_linear.score(X, y))

# RBF kernel (default)
svm_rbf = SVC(kernel='rbf', C=1.0).fit(X, y)
print('RBF SVM score:', svm_rbf.score(X, y))
```

## 2. Margin and kernels

- **Linear:** best when data is roughly linearly separable after scaling.
- **RBF:** maps to infinite-dimensional space; **gamma** controls locality of influence.
- **Polynomial:** interaction terms up to degree `degree`.

**C** penalizes margin violations (smaller margin when C is small, more regularization-like behavior).

```python
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
svm_rbf2 = SVC(kernel='rbf', C=1.0, gamma='scale').fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    svm_rbf2, X_train, response_method='predict', alpha=0.3, ax=ax
)
ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', s=40)
ax.set_title('RBF SVM decision regions (first 2 features)')
plt.show()
print('Test accuracy:', svm_rbf2.score(X_test, y_test))
```
