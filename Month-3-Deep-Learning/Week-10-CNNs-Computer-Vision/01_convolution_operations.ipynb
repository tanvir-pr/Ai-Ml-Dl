# Convolution Operations

## Overview

- Feature maps

**Time:** 2 hours

```python
import numpy as np
from scipy import ndimage

# Simple 2D convolution
img = np.random.rand(5, 5)
filter_3x3 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])  # Edge detect
result = ndimage.convolve(img, filter_3x3, mode='constant', cval=0)
print('Input shape:', img.shape)
print('Output shape:', result.shape)

# Keras Conv2D
import tensorflow as tf
layer = tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1))
print('Conv2D output shape:', layer.output_shape)
```
