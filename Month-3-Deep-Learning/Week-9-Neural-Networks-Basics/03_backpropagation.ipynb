```python
import numpy as np

# Simplified backprop example: dL/dw for single neuron
# L = (y_pred - y)^2, y_pred = sigmoid(w*x + b)
# dL/dw = dL/dy_pred * dy_pred/dz * dz/dw

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Numerical gradient check
w, x, y = 0.5, 1.0, 0.8
z = w * x
pred = sigmoid(z)
loss = (pred - y)**2
dL_dpred = 2 * (pred - y)
dpred_dz = sigmoid_derivative(z)
dz_dw = x
dL_dw = dL_dpred * dpred_dz * dz_dw
print('Analytical gradient:', dL_dw)

# Numerical
eh = 1e-5
loss_plus = (sigmoid((w+eh)*x) - y)**2
loss_minus = (sigmoid((w-eh)*x) - y)**2
numerical = (loss_plus - loss_minus) / (2*eh)
print('Numerical gradient:', numerical)
```
