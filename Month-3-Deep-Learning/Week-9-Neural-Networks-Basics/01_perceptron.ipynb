# Perceptron and Activation Functions

## Overview
- Single neuron (perceptron)
- Activation functions: sigmoid, ReLU, tanh
- Forward pass
- AND/OR gate implementation

**Time:** 1-2 hours

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-5, 5, 100)
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid')
plt.subplot(132)
plt.plot(x, relu(x))
plt.title('ReLU')
plt.subplot(133)
plt.plot(x, np.tanh(x))
plt.title('Tanh')
plt.show()

# AND gate with perceptron
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])
w = np.array([0.5, 0.5])
b = -0.8
pred = (X @ w + b > 0).astype(int)
print('AND gate:', pred)
```
