# MLP from Scratch

## Overview
- Multi-layer perceptron
- Forward propagation
- Backpropagation basics
- NumPy implementation

**Time:** 2-3 hours

```python
import numpy as np

np.random.seed(42)

class SimpleMLP:
    def __init__(self, layer_sizes):
        self.weights = [np.random.randn(a, b) * 0.1 for a, b in zip(layer_sizes[:-1], layer_sizes[1:])]
        self.biases = [np.zeros((1, b)) for b in layer_sizes[1:]]
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def forward(self, X):
        a = X
        for w, b in zip(self.weights[:-1], self.biases[:-1]):
            a = self.relu(a @ w + b)
        return a @ self.weights[-1] + self.biases[-1]

# Test
mlp = SimpleMLP([2, 4, 1])
X = np.random.randn(10, 2)
print('Output shape:', mlp.forward(X).shape)
```
