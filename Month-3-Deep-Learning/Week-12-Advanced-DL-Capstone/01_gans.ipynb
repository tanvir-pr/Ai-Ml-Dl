```python
import tensorflow as tf
# GAN: Generator creates fake, Discriminator tries to detect
# Training: Generator tries to fool Discriminator
# Loss: min_G max_D [log D(x) + log(1-D(G(z)))]

from tensorflow.keras import layers, Model

# Simple generator: noise -> image
def build_generator(latent_dim):
    model = tf.keras.Sequential([
        layers.Dense(256, input_dim=latent_dim),
        layers.LeakyReLU(),
        layers.Dense(512),
        layers.LeakyReLU(),
        layers.Dense(784, activation='tanh')
    ])
    return model

print('GAN: Generator + Discriminator adversarial training')
```
